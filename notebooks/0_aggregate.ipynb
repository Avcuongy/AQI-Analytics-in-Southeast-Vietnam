```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_1 = pd.read_csv("../data/historical/air_quality/air_quality_year_2023.csv")
df_2 = pd.read_csv("../data/historical/air_quality/air_quality_year_2024.csv")
df_3 = pd.read_csv("../data/historical/air_quality/air_quality_year_2025.csv")
df_aqi = pd.concat([df_1, df_2, df_3], axis=0, ignore_index=True)
df_aqi.head()
```

```python
df_aqi.info()
```

```python
df_1 = pd.read_csv("../data/historical/weather/weather_year_2023.csv")
df_2 = pd.read_csv("../data/historical/weather/weather_year_2024.csv")
df_3 = pd.read_csv("../data/historical/weather/weather_year_2025.csv")
df_weather = pd.concat([df_1, df_2, df_3], axis=0, ignore_index=True)
df_weather.head()
```

```python
df_weather.info()
```

```python
df = pd.merge(
    df_aqi, df_weather, on=["date", "location_id", "location_name"], how="inner"
)
```

```python
df.info()
```

```python
df.head()
```

```python
df.to_csv(
    "../data/clean/data.csv",
    index=False,
)
```
